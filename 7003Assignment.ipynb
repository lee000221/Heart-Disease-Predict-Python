### 2.3 Data cleaning


```python
# AgeCategory transfer to int/float for calculation, I set the mean age for each group of age
new_AgeCategory = {'80 or older':82, '75-79':77, '70-74':72, '65-69':67, '60-64':62,'55-59':57, '50-54':52,'45-49':47,'40-44':42,'35-39':37,
                      '30-34':32,'25-29':27, '18-24':22,}
df['AgeCategory'] = df['AgeCategory'].apply(lambda x: new_AgeCategory[x])
df['AgeCategory'] = df['AgeCategory'].astype('int')
df.rename(columns = {"AgeCategory":"MeanAge"},inplace = True)
```

```python
bool_cols = []
# No = 0, Yes = 1 for two unique values columns.
# Female = 0, Male = 1
for col in df.columns:
    if df[col].unique().shape[0] == 2:
      print('Values for columns with 2 unique values')
      print("'{}' unique values: {}".format(col, df[col].unique()))
      bool_cols.append(col)
```

```python
#For Diabetic: No = 0, No, borderline disbetes = 1, Yes = 2, Yes(during pregnancy) = 3
for col in df.columns:
    if df[col].unique().shape[0] == 4:
      print('Values for columns with 4 unique values')
      print("'{}' unique values: {}".format(col, df[col].unique()))
      bool_cols.append(col)
```

```python
#For Genhealth Fair = 0, Very good = 1, Good = 2, Poor = 3, Excellent = 4.
for col in df.columns:
    if df[col].unique().shape[0] == 5:
      print('Values for columns with 5 unique values')
      print("'{}' unique values: {}".format(col, df[col].unique()))
      bool_cols.append(col)
```

```python
# White = 0, Black = 1, Asian = 2, American Indian/Alaskan Native = 3, 'Other' = 4, 'Hispanic' = 5
for col in df.columns:
    if df[col].unique().shape[0] == 6:
      print('Values for columns with 6 unique values')
      print("'{}' unique values: {}".format(col, df[col].unique()))
      bool_cols.append(col)
```

```python
# all yes/no transfer to 1/0
cat_features = df.select_dtypes(include=['object']).columns.tolist()
df[cat_features] = df[cat_features].apply(lambda x: pd.factorize(x)[0])
```

```python
# Keep SleepTime from 4 to 20 only, it actually improve around 1% accuracy compare to drop this column.
df = df[df['SleepTime'] > 3]
df = df[df['SleepTime'] <= 20]
df.head(5)
```

**There are a lot of data cleaning for EDA purpose also include sampling part. The over data cleaning may reduce the performance of the model as well. So, first step of modelling is to reload the dataset and only do the necessary data cleaning.**


```python
# Reload dataset from github.
ds = pd.read_csv('https://raw.githubusercontent.com/lee000221/introdatascience/main/heart_1.csv')

# Re assign AgeCategory as Mean Age of the age distance
new_AgeCategory = {'80 or older':82, '75-79':77, '70-74':72, '65-69':67, '60-64':62,'55-59':57, '50-54':52,'45-49':47,'40-44':42,'35-39':37,
                      '30-34':32,'25-29':27, '18-24':22,}
ds['AgeCategory'] = ds['AgeCategory'].apply(lambda x: new_AgeCategory[x])
ds['AgeCategory'] = ds['AgeCategory'].astype('int')
ds.rename(columns = {"AgeCategory":"MeanAge"},inplace = True)

# Transfer yes/no to 1/0
cat_features = ds.select_dtypes(include=['object']).columns.tolist()
ds[cat_features] = ds[cat_features].apply(lambda x: pd.factorize(x)[0])

# Drop the sleeptime with 1,2,3,21,22,23,24.
ds = ds[ds['SleepTime'] > 3]
ds = ds[ds['SleepTime'] <= 20]

ds.head(5)
```

```python
ds.corr()
```

**Genhealth and sleeptime has very small corr with heart disease, but sleeptime after we processed, corr up to 0.017.**


```python
# Drop gen health column
ds.drop(['GenHealth'],axis=1,inplace=True)
```

**Set HeartDisease as predict y and column except heartdisease as X.**

```python
# Drop heartDisease and put it to y
X = ds.drop('HeartDisease',axis=1)
y = ds['HeartDisease']
print(f' X_shape: {X.shape} \n y_shape: {y.shape}')
```

**Next step is to split dataset as train set (80%) and test set (20%). The train set without Random Over Sampling named as X_train,y_train**

```python
#Split dataset without ROS to train with 80% and 20% as test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,shuffle=True,random_state=0)
print(f' X_train: {X_train.shape} & X_test: {X_test.shape}')
print(f' y_train: {y_train.shape} & y_test: {y_test.shape}')
```

**Do the standardzation**

```python
# Standard original data
from sklearn.preprocessing import StandardScaler
scl = StandardScaler()
X = scl.fit_transform(X)
```

**The Random Over sampler algorithm will be applied on the X_train and y_train and  rename as X_ros_train and y_ros_train.**

```python
# Split dataset with ROS processed to train with 80% and 20% as test
# X_ros_train, X_ros_test, y_ros_train, y_ros_test = train_test_split(X,y,test_size=0.20,shuffle=True,random_state=0)

# Use the RandomOverSampler algorithm to increase the size of training dataset
from imblearn.over_sampling  import RandomOverSampler
ros = RandomOverSampler(sampling_strategy=1)
X_ros_train, y_ros_train = ros.fit_resample(X_train,y_train)

print(f' X ros train: {X_ros_train.shape} & X test: {X_test.shape}')
print(f' y ros train: {y_ros_train.shape} & y test: {y_test.shape}')
```

**Accuracy can be a very good measure of model performance when the data is balanced, but may not be a good measure if the dataset is unbalanced. The dataset used in this paper is a typical case of extremely imbalanced datasets.**

**Therefore, this paper further verifies the effective performance of the model through precision, recall, F1score, and confusion matrix to increase the persuasiveness.**

**The precision, recall, and F1 score here are divided into performance when the target is 0 and performance when the target is 1. Therefore, when screening models, the performance of the target as 1, that is, the prediction result of TP, will be used as the model selection criterion.**

**The precision here refers to the ratio of how many TPs are found in the cases where P is found (TP+FP). Recall refers to the proportion of TP found in all cases (TP+FN). If a model has high precision but low recall, it can be understood that there are a large number of FNs in this model. That is, a large number of heart disease patients are judged by the model as normal people. So even if the precision of this model is high, it is not desirable. Conversely, if a model has low precision but high recall, it can be understood that the model has high FP. That is, the probability of judging normal people as heart disease patients. The F1 score is a model quality parameter that comprehensively considers the model precision and recall. Therefore, the criteria for selecting a model can be very clear. The first thing to consider is the better model with higher recall, followed by accuracy and precision.**

image.png

**As shown in the above compairation table, XGB, LR, and Bagging have the highest Recall. At the same time, their precision is not much different. The Voting and Random Forest models have the highest precision. However, their Recall is very low. It means that they have judged a large number of heart patients as normal people. So these two models are not preferred. Therefore, we finally concluded that the bagging model with the lowest accuracy but the highest recall was selected as the best performing model after ROS training.**
